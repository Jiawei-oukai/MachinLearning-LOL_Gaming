```python
import pandas as pd
```

```python
df = pd.read_csv("../processed_data/train.csv")
```

```python
# define useless columns
drop_columns = ['gameId']

prefix = "Player_"
suffixs =["_pick","_position","_time_game"]
for i in range(1,11):
    for suffix in suffixs:
        drop_columns.append(prefix+str(i)+suffix)

drop_columns
```

```python
df.drop(drop_columns, axis=1, inplace=True)
```

```python
df = df.rename(columns={'teamVictory':'TeamVictory'})
```
